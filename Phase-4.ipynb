```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame(np.random.rand(3, 3), columns=['X', 'Y', 'Z'])
df3 = pd.DataFrame(np.zeros((4, 4)), columns=['C1', 'C2', 'C3', 'C4'])
```

```python
df1.to_csv('table1.csv', index=False)
df2.to_json('table2.json', orient='records', indent=2)
df3.to_pickle('table3.pkl')
```

```python
loaded_table_1 = pd.read_csv('table1.csv')
print(loaded_table_1)
```

```python
try:
    sales = pd.DataFrame(
        {
            'category': ['biryani', 'chinese', 'pizza', 'burger'],
            'sales': [120, 95, 140, 110],
        }
    )
    sales.to_csv('pandas-sales.csv', index=False)

    loaded_sales = pd.read_csv('pandas-sales.csv')

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.bar(loaded_sales['category'], loaded_sales['sales'])
    plt.title('Pandas Sales Chart')
    plt.grid(False)

    growth = loaded_sales['sales'] * 1.1
    plt.subplot(122)
    plt.bar(loaded_sales['category'], growth)
    plt.title('Projected Sales (+10%)')
    plt.grid(False)

except FileNotFoundError:
    print('Pandas data file not found')
```
