## Advance operation with Business example


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = pd.DataFrame(
    {
        'restaurant_id': [1, 2, 3, 4, 5],
        'restaurant_name': ['Paradise Biryani', 'Beijing Bites', 'Pizza Hub', 'Burger Point', 'Chai Point'],
        'restaurant_type': ['biryani', 'chinese', 'pizza', 'burger', 'tea'],
        '2021': [150000, 120000, 200000, 180000, 160000],
        '2022': [180000, 140000, 230000, 210000, 185000],
        '2023': [220000, 160000, 260000, 240000, 205000],
        '2024': [250000, 190000, 300000, 270000, 230000],
    }
)

print("=== Zomato Sales Analysis with Pandas ===")
print(f"Sales Data Shape: {sales_data.shape}")
print(f"Sample data for 1st 3 restaurants:{sales_data.head(3)}")
```

```python
year_cols = ['2021', '2022', '2023', '2024']

yearly_total = sales_data[year_cols].sum(axis=0)
print(f"Yearly total sales:{yearly_total}")

min_sales = sales_data[year_cols].min(axis=1)
print(f"Minimum sales per restaurant:{min_sales}")

max_sales = sales_data[year_cols].max(axis=0)
print(f"Maximum sales per year:{max_sales}")

avg_sales = sales_data[year_cols].mean(axis=1)
print(f"Average sales per restaurant:{avg_sales}")
```

```python
year_cols = ['2021', '2022', '2023', '2024']
cumsum = sales_data[year_cols].cumsum(axis=1)
print(f"Cumulative sales:{cumsum}")

plt.figure(figsize=(8, 6))
plt.plot(cumsum.mean(axis=0))
plt.title('Average cumulative sales across all restaurant')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
sales_2023 = sales_data['2023'].to_numpy()
sales_2024 = sales_data['2024'].to_numpy()

print(f"Addition of 2023 and 2024 sales vectors: {sales_2023 + sales_2024}")
print(f"Multiplication of vectors: {sales_2023 * sales_2024}")
print(f"Dot Product: {np.dot(sales_2023, sales_2024)}")

angle = np.arccos(
    np.dot(sales_2023, sales_2024)
    / (np.linalg.norm(sales_2023) * np.linalg.norm(sales_2024))
)
print(angle)
```

```python
vectorized_upper = sales_data['restaurant_type'].str.upper()
print(f"Vectorized Upper:{vectorized_upper}")
```
