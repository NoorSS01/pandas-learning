## Pandas DataFrame and basics


```python
import pandas as pd
import numpy as np
```

Creating Series and DataFrame from List and Dictionary


```python
series_1d = pd.Series([1, 2, 3], name='numbers')
print(f"1D Series:{series_1d}")

df_2d = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(f"2D DataFrame:{df_2d}")
```

### List vs Pandas Series


```python
py_list = [1, 2, 3]
print(f"Python list Multiplication: {py_list * 2}")

pd_series = pd.Series([1, 2, 3])
print(f"Pandas Series Multiplication: {list(pd_series * 2)}")

import time
start = time.time()
py_list = [i * 2 for i in range(1000000)]
print("List operation time:", time.time() - start)

start = time.time()
pd_series = pd.Series(np.arange(1000000)) * 2
print("Pandas operation time:", time.time() - start)
```

### creating DataFrames from scratch


```python
zeros_df = pd.DataFrame(np.zeros((3, 4)))
print(f"zeros DataFrame:{zeros_df}")

ones_df = pd.DataFrame(np.ones((2, 3)))
print(f"ones DataFrame:{ones_df}")

full_df = pd.DataFrame(np.full((4, 4), 7))
print(f"full DataFrame:{full_df}")

random_df = pd.DataFrame(np.random.random((3, 4)))
print(f"random DataFrame:{random_df}")

sequence_series = pd.Series(np.arange(0, 10, 2))
print(f"sequence Series:{sequence_series}")
```

### Series and DataFrame


```python
vector_series = pd.Series([1, 2, 3], name='vector')
print(f"Series:{vector_series}")

matrix_df = pd.DataFrame([[1, 2, 3], [1, 2, 4]], columns=['A', 'B', 'C'])
print(f"DataFrame:{matrix_df}")

tensor_like = pd.DataFrame(
{
        'store': ['A', 'B'],
        'Q1': [[100, 120], [140, 160]],
        'Q2': [[180, 200], [210, 230]],
    }
)
print(f"Nested DataFrame (tensor-like values):{tensor_like}")
```

### DataFrame Properties


```python
arr_df = pd.DataFrame([[1, 2, 3], [4, 5, 6]], columns=['X', 'Y', 'Z'])
print(f"Shape: {arr_df.shape}")
print(f"Dimension: {arr_df.ndim}")
print(f"Size: {arr_df.size}")
print(f"DTypes:{arr_df.dtypes}")
```

```python
df = pd.DataFrame(np.arange(12).reshape(3, 4), columns=['A', 'B', 'C', 'D'])
print(f"Original DataFrame:{df}")

stacked = df.stack()
print(f"Stacked (flattened view):{stacked}")

unstacked = stacked.unstack()
print(f"Unstacked DataFrame:{unstacked}")

transpose_df = df.T
print(f"Transpose DataFrame:{transpose_df}")
```
