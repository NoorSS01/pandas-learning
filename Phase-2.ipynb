## Pandas DataFrame Operations


```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        'name': ['Aarav', 'Diya', 'Rohan', 'Meera', 'Kabir', 'Anaya', 'Vihaan', 'Ira', 'Reyansh', 'Sara'],
        'age': [21, 19, 24, 22, 26, 20, 23, 25, 28, 18],
        'salary': [45000, 52000, 61000, 58000, 72000, 49000, 65000, 70000, 85000, 47000],
    }
)
print(f"Basic Slicing:{df.iloc[2:7]}")
print(f"With Step:{df.iloc[1:8:2]}")
print(f"Negative index row:{df.iloc[-3]}")
```

```python
print(f"Specific element (row 1, salary): {df.iloc[1, 2]}")
print(f"Entire row 1:{df.iloc[1]}")
print(f"Entire column salary:{df['salary']}")
```

## Sorting


```python
unsorted_df = pd.DataFrame(
    {
        'item': ['pen', 'book', 'eraser', 'pencil', 'scale'],
        'price': [30, 100, 20, 15, 25],
    }
)
print(f"Sorted DataFrame by price:{unsorted_df.sort_values('price')}")

score_df = pd.DataFrame({'A': [3, 1, 2], 'B': [1, 2, 3]})
print(f"Sorted by column A:{score_df.sort_values('A')}")
```

## Filter


```python
even_age = df[df['age'] % 2 == 0]
print(f"Rows with even age:{even_age}")
```

## Filters with Mask


```python
mask = df['salary'] > 60000
print(f"Rows with salary greater than 60000:{df[mask]}")
```

## Fancy indexing vs query()


```python
indices = [0, 2, 4, 7]
print(df.iloc[indices])

query_results = df.query('salary > 60000')
print(f"Rows from query:{query_results}")
```

```python
df['salary_band'] = np.where(df['salary'] > 60000, 'High', 'Regular')
print(df[['name', 'salary', 'salary_band']])
```

## Adding and Removing data


```python
part1 = pd.DataFrame({'id': [1, 2, 3], 'score': [60, 70, 80]})
part2 = pd.DataFrame({'id': [4, 5, 6], 'score': [75, 85, 90]})

combined = pd.concat([part1, part2], ignore_index=True)
print(combined)
```

## DataFrame Compatibility


```python
a = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
b = pd.DataFrame({'A': [4, 5, 6, 7], 'B': [8, 9, 10, 11]})
c = pd.DataFrame({'A': [7, 8, 9], 'B': [10, 11, 12]})
print(f"Compatibility check: {a.shape == b.shape}")
print(f"Compatibility check with c: {a.shape == c.shape}")
```

```python
original = pd.DataFrame({'A': [1, 3], 'B': [2, 4]})
new_row = pd.DataFrame({'A': [5], 'B': [6]})

with_new_row = pd.concat([original, new_row], ignore_index=True)
print(original)
print(with_new_row)

original['C'] = [5, 5]
print(original)
```

```python
arr = pd.DataFrame({'values': [1, 2, 3, 4, 5]})
deleted = arr.drop(index=2)
print(f"DataFrame after deletion:{deleted}")
```
